Load Environment

```python
from dotenv import load_dotenv
import os

load_dotenv("../../.env")
```

OpenAI Client

```python
import openai
client = openai.OpenAI(
    base_url="https://api.cerebras.ai/v1",
    api_key=os.environ.get("CEREBRAS_API_KEY")
)

response = client.chat.completions.create(
    model="gpt-oss-120b",
    messages=[
        {
            "role":"system",
            "content": "You are a helpful assistant that can answer questions and help with tasks."
        },
        {
            "role": "user",
            "content": "Hello, how are you?"
        }
    ]
)
print(response.choices[0].message.content)
```

```python
response.usage
```

Google Genai

```python
from google import genai
genai_client = genai.Client(api_key=os.environ.get("GOOGLE_API_KEY"))

response = genai_client.models.generate_content(model="gemini-2.5-flash", contents="Hello, how are you?")
print(response.text)
```

Groq

```python
from groq import Groq
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))
chat_completion = client.chat.completions.create(model="llama3-8b-8192", messages=[{"role": "user", "content": "Hello, how are you?"}])
print(chat_completion.choices[0].message.content)
```
